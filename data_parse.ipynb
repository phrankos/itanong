```python
import json
import os
```

```python
def format_data(index):
    #open llama3-gendata.json
    new_data = []
    with open(f'llama3-gendata{index}.json') as f:
        data = json.load(f)
        eos = ["\n","<data_start>", "<data_end>", "<eos>", "<info_start>","<info_end>"]
        for item in data:
            for i in eos:
                item = item.replace(i, "")
                #remove all multiple spaces
                item = " ".join(item.split())
            #write to new json file
            #find index of first curly brace
            start = item.find("[")
            #find index of last curly brace
            end = item.rfind("]")
            new_data.append(item[start:end+1])
            
        print(f"total_data: {len(new_data)}")

                    
        with open(f'llama3-new-data{index}.json', 'w') as file:
            file.write("[")
            for data in new_data:
                file.write(data + ",")
                file.write("\n")
            file.write("]")
            file.close()
        f.close()
```

```python
indeces = [0]

for index in indeces:
    format_data(index)
    print(f"llama3-new-data{index}.jsonl created")
```

```python
def synth_data(index):
    
    format_data = []
    #open new json file
    with open(f'llama3-new-data{index}.json') as f:
        try:
            data = json.load(f)
            for i in data:
                format_data.append(i)
        except Exception as e:
            print(e,f)
            
            
    #write formatted data to new file
    print(f"Number of new data: {len(format_data)}")
    counter = 0
    with open(f'llama3-formatted-data{index}.jsonl', 'w') as file:
        for item in format_data:
            print(type(item))
            for lines in item:
                print(type(lines))
                try:
                    file.write(json.dumps(lines))
                    file.write("\n")
                    counter += 1
                except Exception as e:
                    print(e)
                
    print(f"Number of formatted data: {counter}")
```

```python
indeces1 = [0]
for index in indeces1:
    synth_data(index)
    print(f"llama3-formatted-data{index}.jsonl created")
```
